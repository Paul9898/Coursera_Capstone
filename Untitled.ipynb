## Modelling

```python
X = df_new.drop(['SEVERITYCODE'],axis=1)
X.head()
```

```python
y = df_new['SEVERITYCODE'].astype(np.int64).astype('str').values
y[0:5]
```

```python
X= preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn import metrics
```

## Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
Tree = DecisionTreeClassifier(criterion="entropy", max_depth = 5)
Tree # it shows the default parameters
```

```python
Tree.fit(X_train,y_train)
predTree = Tree.predict(X_test)
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
```

```python
round(f1_score(y_test, predTree, average='weighted'),4)
```

## SVM

```python
from sklearn import svm
SV = svm.SVC(kernel='rbf')
SV.fit(X_train, y_train)
yhat = SV.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
cnf_matrix = confusion_matrix(y_test, yhat, labels=['1','2'])
np.set_printoptions(precision=2)

print (classification_report(y_test, yhat))
```

```python
print('SVM\'s Accuracy: ',np.sum(y_test==yhat)/len(yhat))
```

```python
round(f1_score(y_test, yhat, average='weighted'),4)
```

## Logistic

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat_prob = LR.predict_proba(X_test)
```

```python
print('Logistic Regression\'s Accuracy: ',np.sum(y_test==yhat)/len(yhat))
```

```python
round(f1_score(y_test, yhat, average='weighted'),4)
```

```python
round(log_loss(y_test, yhat_prob),4)
```
